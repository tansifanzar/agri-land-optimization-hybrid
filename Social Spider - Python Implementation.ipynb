```python
dij=np.zeros((10,10))
distance=np.zeros(90)
dist=np.zeros(100)
d=0
d1=0

for i in range (n):
    for j in range (n):
        dij[i][j]=round(math.sqrt(math.pow(weight[i] - weight[j], 2)+ math.pow(TLA[i] - TLA[j], 2)),3)
        dist[d1]=dij[i][j]
        d1=d1+1
        if (i!=j):
            dij[i][j]=round(math.sqrt(math.pow(weight[i] - weight[j], 2) + math.pow(TLA[i] - TLA[j], 2)), 3)
            distance[d]=dij[i][j]
            print("Distance between spiders ",i," and ",j," is: ",dij[i][j])
            d=d+1
mindis=distance.min()
```

```python
male=np.zeros((10,10))
number1 = 0
number2 = 0
wfind = str(np.where(dist==mindis)[0][0])
numberArray1 = np.zeros(len(wfind))
counter1=0
mo=np.zeros(100)
m1=0
k2=0
i3=0
nuf1=int(Nf)

for i1 in range (len(wfind)):
    numberArray1[i1]=wfind[counter1:counter1+1]
    counter1=counter1+1
if (len(wfind)==1):
    number1=0
    number2=int(numberArray1[0])
elif (len(wfind)==2):
    number1=int(numberArray1[0])
    number2=int(numberArray1[1])
    
wf=weight[number2]

for m1 in range (Nm):
    if (m1<Nf & number2<Nf):
        dis=math.sqrt(math.pow((TLA[m1]-TLA[number2]),2)+math.pow((weight[m1]-weight[number2]),2))
        expo11=math.pow(dis,2)
        expol2=math.exp(-expo11)
        exp12=wf*expol2
        vibfi=round(exp12,3)
        print("Vibfi values is ",vibfi)
        
    for k2 in range (9):
        if (mweight[m1]>mweight[int(nfm)]):
            male[m1][k2+1]=male[m1][k2]+a1*vibfi*(wf-male[m1][k2])+g1*(r1-0.5)
            mo[i3]=round(male[m1,k2+1],3)
            print("Male cooperative of ",m1," and ",k2," is ",mo[i3])
            i3=i3+1
        else:
            for h in range (1,Nm):
                wnf=weight[nuf1+h]
                nr=male[h,k2]*wnf
                nrt=nr+nrt
                drt=wnf
                drt1=drt+drt1
            male[m1][k2+1]=male[m1][k2]+a1*(nrt/drt1-male[m1][k2])
            mo[i3]= round(male[m1][k2+1],3)
            print("Male cooperative of ",m1," and ",k2," is ",mo[i3])
            i3=i3+1
```
